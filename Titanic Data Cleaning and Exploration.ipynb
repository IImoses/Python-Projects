```python
#lets change the survived column into died and survived
new_survived =  pd.Categorical(df["Survived"])
new_survived = new_survived.rename_categories(["Died","Survived"])
df["Survived"] = new_survived
df.head(6)
```

```python
#lets change the pclass column into different classes to show 1st class 2nd class and 3rd class
new_Pclass =  pd.Categorical(df["Pclass"], ordered = True)
new_Pclass = new_Pclass.rename_categories(["Class1","Class2", "Class3"])
df["Pclass"] = new_Pclass
df.head(15)
```

```python
#lets look at cabin column. we want to make the cabin column contain only the first letter of the string in each cell to remove the randomness
char_cabin = df["Cabin"].astype(str) #to convert the column to string type
new_cabin = np.array([cabin[0] for cabin in char_cabin]) #take the first leter
new_cabin = pd.Categorical(new_cabin)
new_cabin.describe()
df["Cabin"] = new_cabin
```

```python
#we want to find out how to populate the column. lets see if the data follows a trend so we can use the median value to fill the missing values
df.Age.hist(figsize=(6,4))
```

```python
#the data follows a trend so we'll use the median
df.Age.median()
```

```python
#use the numpy function to make an array of a variable in which where Age column is null,28 is inputed, else the value remains the original value
null_ages = np.where( df["Age"].isnull(), 28, df["Age"]) 
df.Age = null_ages
df.Age.describe()
#now count of age column shows it is complete
```

```python
#we want to find out what the fare column looks like. ie the distribution
df.boxplot("Fare")
```

```python
#lets see the max fare values

max_fare = np.where(df["Fare"] == max(df["Fare"]))
df.loc[max_fare]
```

```python
#sibsp and parch columns show how many siblings and parents respectively each person had onboard.
#we can create a new column to just show how many family members each person had ie sibligs + parents

df["Family"] = df["SibSp"] + df["Parch"]
df.head(5)
```
